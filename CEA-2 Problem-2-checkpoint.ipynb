```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#reading file into data frame
df = pd.read_csv('ex1data2.txt',header = None)
print(df)
```

```python
#converting it into numpy array
df = np.array(df)
x = df[::,0:2] # all rows and column 0 to 2
y = df[::,2:3] # all rows and column 2 to 3
print(x.shape)
print(y.shape)
```

```python
#plotting graph of the given data
plt.plot(x,y,color = 'g')
plt.show()
```

```python
#now adding a column of 1's to the x to make its inverse possible
x1_s = np.ones((47,1))
x1_s.shape
```

```python
#horizontally stacking the column of 1's
x = np.hstack((x,x1_s))
x.shape
```

```python
#taking transpose and multiplying it with x
tran_x = np.transpose(x)
dot_x = np.dot(tran_x,x)
dot_x.shape
```

```python
#taking inverse of dot_x
xn = np.linalg.inv(dot_x)
xn.shape
```

```python
x2= np.dot(xn,tran_x)
x2.shape
```

```python
Q = np.dot(x2,y)
print(Q)
```

```python
Q0=Q[0]
Q1 = Q[1]
Q2 = Q[2]
print(Q1)
print(Q2)
```

```python
from tkinter import *
```

```python
caed = Tk()
caed.configure(bg = 'purple')
caed.title("House Price Prediction System")
caed.geometry("480x130")
caed.minsize(450,130)
caed.maxsize(800,800)
def func():
    profit = Q1* n1.get() + Q2* n2.get() + Q0
    
    t1.delete("1.0",END)
    t1.insert(END,profit)
    
    
t2 = Label(caed,text = "Enter size of house (in square feet)")
t2i = Label(caed,text = "Enter number of bedrooms")

n1 = DoubleVar()
n2 = IntVar()
t4 = Entry(caed,textvariable = n1)
t4i = Entry(caed,textvariable = n2)
t5 = Label(caed,text = "Price")

t1 = Text(caed,height = 1,width = 20)
b = Button(caed,text = "Press",command = func)
b.configure(bg = 'grey')

t2.grid(row = 0,column = 0,padx = 1)
t2i.grid(row = 1,column = 0)
t4.grid(row = 0,column = 2,pady = 2)
t4i.grid(row = 1,column = 2,pady = 2)

t5.grid(row = 3,column = 1,pady = 2)
t1.grid(row = 4,column = 1,pady = 2)
b.grid(row = 2,column = 2)


caed.mainloop()
```
